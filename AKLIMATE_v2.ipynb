```python
count_dict={} # 104 subtype keys, value for each counting number of occurances
correct_dict={} # 104 subtype keys, value for each counting number correct predictions

cncr_typs=['ACC','BLCA','BRCA','CESC','COADREAD',
            'ESCC','GEA','HNSC','KIRCKICH','KIRP',
            'LGGGBM','LIHCCHOL','LUAD','LUSC','MESO',
            'OV','PAAD','PCPG','PRAD','SARC',
            'SKCM','TGCT','THCA','THYM','UCEC',
            'UVM']

#cncr_typs=['ACC']#devel
for c_typ in cncr_typs:
    file=pd.read_csv('predictions_collected_top_50_ranked_feature_importance_'+c_typ+'_20200417.tsv',
                delimiter='\t', header=[1]) #picking correct double header in tsv
        
    for row in file.index: # Go through each row in cancer file n
        label=file.iloc[row, 4] # Get ground truth
        x=file.iloc[row, 5] # Get prediction info json obj
        y=json.loads(x)     # Convert to dict of dicts
        z=y['classification'] # Extract nested dict of scores
        max_id=max(z.items(), key=operator.itemgetter(1))[0] # Get id max score
        
        if not label in count_dict:
            count_dict[label]=0
        if not label in correct_dict:
            correct_dict[label]=0
        count_dict[label]=count_dict[label]+1
        if label == max_id:
            correct_dict[label]=correct_dict[label]+1
cols=['Label_AKLI','Score_AKLI']
result_akli=pd.DataFrame(columns=cols,index=list(range(0,104)))
for i, label in enumerate(count_dict):
    result_akli.iloc[i,0]=label
    result_akli.iloc[i,1]=correct_dict[label]/count_dict[label]
print(result_akli)
```
