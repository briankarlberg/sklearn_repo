```python
count_dict={}
correct_dict={}
count=0
header = None
#with gzip.open("2020-03-20-allCOHORTS_20200203Tarball_JasleenGrewal.gz", "rt", encoding='utf-8') as handle:
with open("2020-03-20-allCOHORTS_20200203Tarball_JasleenGrewal", "rt", encoding='utf-8') as handle:
    for line in handle:
        count=count+1
#         if count > 1000:
#             break
        if not line.startswith("#"):
            row = line.rstrip().split("\t") # row object is a list
            if header is None:
                header = row # first time through loop
            else:
                sample = row[0]
                repeat = row[1]
                fold = row[2]
                test = row[3]
                label = row[4]
                meta = json.loads(row[5].replace("'", "\""))
                cls = pandas.Series(meta['classification'])
                c = cls.idxmax()
                #print("%s == %s" % (label, c))
                if not label in count_dict:
                    count_dict[label]=0
                if not label in correct_dict:
                    correct_dict[label]=0
                count_dict[label]=count_dict[label]+1
                if label == c:
                    correct_dict[label]=correct_dict[label]+1
# print(count_dict)  
# print(correct_dict)
for label in count_dict:
    print (label+' accuracy: '+str(correct_dict[label]/count_dict[label]))
```
